```python
class MyName:
    """Class for creating name instances with email and ID."""
    total_names = 0  # Class Variable

    def __init__(self, name=None) -> None:
        self.name = name if name is not None else self.anonymous_user().name  # Class attributes / Instance variables
        MyName.total_names += 1  # Modify class variable
        self.my_id = MyName.total_names

    @property
    def whoami(self):
        """Returns the name of the instance."""
        return f"My name is {self.name}"
    
    @property
    def my_email(self) -> str:
        """Returns an email based on the instance's name."""
        return self.create_email()
    
    def create_email(self) -> str:
        """Generates an email address for the instance."""
        return f"{self.name}@itcollege.lviv.ua"

    @classmethod
    def anonymous_user(cls):
        """Returns an instance of an anonymous user."""
        return cls("Anonymous")
    
    @staticmethod
    def say_hello(message="Hello to everyone!"):
        """Returns a greeting message."""
        return f"You say: {message}"


print("Let's Start!\n")

# Updated names tuple to include "Yarroslav"
names = ("Bohdan", "Marta", "Yarroslav", None)
all_names = {name: MyName(name) for name in names}

for name, me in all_names.items():
    print(f"{'>*< '*10}")
    print(f"This is object: {me}")
    print(f"Object attribute: name = {me.name}, ID = {me.my_id}")
    print(f"{type(MyName.whoami)}: {me.whoami}")
    print(f"Email: {me.my_email}")
    print(f"Static method: {me.say_hello()}")
    print(f"Class variable total_names: {MyName.total_names}")
    print(f"{'<*>'*10}\n")

print(f"We are done. We created {MyName.total_names} names!")
```
